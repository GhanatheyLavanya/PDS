```python
#Reproducible step-3: Data analysis of frailty dataset
#importing librariers
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

## Reading the cleaned data
df_frailty = pd.read_csv("../data_clean/clean_frailty_data.csv")


# Plot the relationship between Frailty and Grip Strength:box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='frailty', y='grip_strength(kg)', data=df_frailty)
plt.xlabel('Frailty')
plt.ylabel('Grip Strength (kg)')
plt.title('Boxplot: Frailty vs. Grip Strength')
plt.grid(True)
plt.show()
            
# Correlation Matrix
correlation_matrix = df_frailty.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
